```python
print(" Attendance Analyzer\n")

# Input total classes
total = int(input("Enter total number of classes: "))

# Input attended classes
attended = int(input("Enter number of classes attended: "))

# Calculate percentage
percentage = (attended / total) * 100

print("\nYour attendance is:", percentage, "%")

# Check eligibility
if percentage >= 75:
    print(" You are eligible for exams")
else:
    print(" You are NOT eligible for exams")

```
